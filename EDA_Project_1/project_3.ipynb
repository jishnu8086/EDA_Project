# Bank Customer Transaction Data

This dataset contains bank transaction records with 2,512 entries and 16 columns. It includes TransactionID, AccountID, TransactionAmount, Date, Type, Location, DeviceID, IP Address, MerchantID, and Channel. Additional details such as CustomerAge, Occupation, Transaction Duration, Login Attempts, Account Balance, and Previous Transaction Date are provided. The dataset helps analyze banking behavior, transaction patterns, and potential fraud detection. Transactions are categorized into credit and debit types and occur across multiple channels like ATMs and online platforms. â€‹

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
df=pd.read_csv(r"C:\datascience\dataset1/bank.csv")
print(df)
print(df.head(10))
```

```python
print(df.info())
```

```python
print(df.isna().sum())
```

```python
df ['TransactionDate']=pd.to_datetime(df['TransactionDate'])
print(df.info())
                                      
```

```python
df.duplicated().sum()
df.drop_duplicates(inplace=True)
```

```python
sns.boxplot(data=df)
plt.show()
```

```python
print(df.rename(columns={'TransactionID':'Transaction_ID','AccountID':'Account_ID','TransactionAmount':'Transaction_Amount','TransactionDate':'Transaction_Date','TransactionType':'Transaction_Type','DeviceID':'Device_ID','IP Address':'IP_Address','Merchant_ID':'Merchant_ID','CustomerAge':'Customer_Age','CustomerOccupation':'Customer_Occupation','TransactionDuration':'Transaction_Duration','LoginAttempts':'Login_Attempts','AccountBalance':'Account_Balance','PreviousTransactionDate':'Previous_Transaction_Date'}))
```

```python
df.describe()
df.describe(include=object)
```

```python
print(df.corr())
```

```python
f=df['TransactionAmount'].corr(df['CustomerAge'])
print(f)
plt.scatter(df['TransactionAmount'],df['CustomerAge'])
plt.title("Correlation Matrix")
plt.xlabel('TransactionAmount')
plt.ylabel('CustomerAge')
plt.show()
```

```python
#highest average transaction amount of top 10 account
avg_transaction = df.groupby("AccountID")["TransactionAmount"].mean().sort_values(ascending=False).head(10)
plt.figure(figsize=(12, 6))
avg_transaction.plot(kind='bar', color='orange')
plt.xlabel("Account ID")
plt.ylabel("Average Transaction Amount")
plt.title("Top 10 Accounts with Highest Average Transaction Amount")
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
#most common transaction type
transaction_counts=df['TransactionType'].value_counts()
most_common_transaction=transaction_counts.idxmax()
most_common_count=transaction_counts.max()
print(f"the most common transaction type is '{most_common_transaction}'with{most_common_count}occurrences.")
```

```python
#accounts with consistently have large balance
large_balance_threshold = 15000
consistency_threshold=4
high_balance_counts=df[df['AccountBalance']>large_balance_threshold].groupby('AccountID')['TransactionDate'].count()
consistent_accounts = high_balance_counts[high_balance_counts >=consistency_threshold]
print(f"total accounts with consistently high balances: {len(consistent_accounts)}")
```

```python
#average number of transaction per account
transactions_per_account = df["AccountID"].value_counts()
average_transactions = transactions_per_account.mean()
top_accounts = transactions_per_account.head(10)

plt.figure(figsize=(12, 6))
top_accounts.plot(kind='bar', color='purple')
plt.xlabel("Account ID")
plt.ylabel("Number of Transactions")
plt.title(f"Top 10 Accounts with Most Transactions (Avg: {average_transactions:.2f})")
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()

```

```python
#mostly used channel for transaction
channel_counts=df['Channel'].value_counts()
print("Transaction counts by Channel:")
print(channel_counts)
plt.figure(figsize=(8,5))
channel_counts.plot(kind='bar',color=['blue','orange','green'])
plt.title("Transactions by Channel")
plt.xlabel("Channel")
plt.ylabel("Transaction Count")
plt.xticks(rotation=45)
plt.show()
```

```python
#average transaction amount per customer
average_transaction_per_customer = df.groupby('AccountID')['TransactionAmount'].mean()
print("Average transaction amount:")
print(average_transaction_per_customer)
plt.figure(figsize=(8,5))
average_transaction_per_customer.plot(kind='bar',color=['red'])
plt.title("Average Transaction Amount ")
plt.xlabel("AccountID")
plt.ylabel("Average Transaction Amount")
plt.xticks(rotation=45)
plt.show()
```

```python
#account with higher average transaction
avg_transaction_per_account= df.groupby('AccountID')['TransactionAmount'].mean()
overall_avg = df['TransactionAmount'].mean()
high_avg_accounts = avg_transaction_per_account[avg_transaction_per_account > overall_avg]
print("Accounts with higher than average transaction amounts :")
print(high_avg_accounts)
plt.title("Accounts with higher than Average Transaction Amount")
plt.xlabel("AccountID")
plt.ylabel("Average Transaction Amount")
plt.xticks(rotation=45)
plt.axhline(overall_avg,color='black',linestyle='dashed',label='Overall Average')
plt.legend()
plt.show()
```

```python


# most transaction of days
df['TransactionDate'] = pd.to_datetime(df['TransactionDate'])
df['day_of_week'] = df['TransactionDate'].dt.day_name()
df['hour_of_day'] = df['TransactionDate'].dt.hour
transactions_by_day = df['day_of_week'].value_counts()

transactions_by_hour = df['hour_of_day'].value_counts().sort_index()
print("Transactions by day of the week:")
print(transactions_by_day)

print("\nTransactions by hour of the day:")
print(transactions_by_hour)
fig, axes = plt.subplots(1, 2, figsize=(12, 5))
transactions_by_day.plot(kind='bar', ax=axes[0], color='skyblue')
axes[0].set_title("Transactions by Day of the Week")
axes[0].set_xlabel("Day")
axes[0].set_ylabel("Transaction Count")
axes[0].set_xticklabels(transactions_by_day.index, rotation=45)
transactions_by_hour.plot(kind='line', ax=axes[1], marker='o', color='red')
axes[1].set_title("Transactions by Hour of the Day")
axes[1].set_xlabel("Hour")
axes[1].set_ylabel("Transaction Count")
axes[1].set_xticks(range(0, 24))
plt.tight_layout()
plt.show()
```

```python
#highest transaction amount with location

totals = df.groupby("Location")["TransactionAmount"].mean()
highest_location = totals.idxmax()
highest_amount = totals.max()
print(f"Location with the highest transaction amount: {highest_location} (${highest_amount})")
```

```python


# highest duration of transaction
df['TransactionDate'] = pd.to_datetime(df['TransactionDate'])
df['PreviousTransactionDate'] = pd.to_datetime(df['PreviousTransactionDate'])
df['duration'] = (df['PreviousTransactionDate'] - df['TransactionDate']).dt.total_seconds()
highest_duration_transaction = df.loc[df['duration'].idxmax()]
print("Transaction with the highest duration:")
print(highest_duration_transaction)
top_longest_transactions = df.nlargest(5, 'duration')
print("\nTop 5 longest transactions:")
print(top_longest_transactions)
```

```python
#customers having same occupation
same_occupation = df.groupby('CustomerOccupation')['AccountID'].apply(list).reset_index()
print(same_occupation)
```

```python
#customers with suspiciously high transaction 
threshold = df['TransactionAmount'].quantile(0.95)
suspicious_transactions = df[df['TransactionAmount']> threshold]
suspicious_account= suspicious_transactions['AccountID'].unique()
print("Suspicious Account IDs:",suspicious_account)
print("\nSuspicious Transaction:")
print(suspicious_transactions)
```

```python
#relationship and correlation coefficient
plt.figure(figsize=(10,5))
sns.regplot(x=df["CustomerAge"],y=df["AccountBalance"],scatter_kws={"alpha":0.5},line_kws={"color":"red"})
plt.xlabel("CustomerAge")
plt.ylabel("AccountBalance")
plt.title("Relationship between CustomerAge and AccountBalance")
plt.show()
correlation=df["CustomerAge"].corr(df["AccountBalance"])
print(f"correlation coefficient:{correlation:.2f}")
```

```python
#transaction from multiple location with same device id
device_location_counts=df.groupby('DeviceID')['Location'].nunique()
suspicious_devices=device_location_counts[device_location_counts > 1]
print(suspicious_devices)
```

```python
#highest transaction duration of an account
highest_duration = df.groupby("AccountID")["TransactionDuration"].max().sort_values(ascending=False).head(10)
plt.figure(figsize=(12, 6))
highest_duration.plot(kind='bar', color='skyblue')
plt.xlabel("Account ID")
plt.ylabel("Highest Transaction Duration")
plt.title("Top 10 Accounts with Highest Transaction Duration")
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

The average transaction duration is 119.64 seconds, ranging from 10 to 300 seconds, which could help optimize service efficiency, especially for online and ATM transactions. Previous transactions are recorded across 360 different timestamps, providing an opportunity to track trends over time and identify spending behaviors.

Overall, the dataset is comprehensive and well-structured, with no missing values, making it valuable for further analysis.
