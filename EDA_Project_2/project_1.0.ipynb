# Football Match of Home and Away Team Data

This dataset contains 4,070 football matches with 33 columns covering match statistics. It includes season, date, home and away teams, goals scored, and match results. Detailed stats such as possession, passes, shots, tackles, fouls, yellow/red cards, and offsides are recorded. The dataset helps analyze team performance, match trends, and player behavior. Matches span different seasons and leagues, allowing comparisons over time. It can be used for predicting match outcomes, performance evaluation, and tactical insights in football analytics.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
df=pd.read_csv(r"C:\datascience\dataset1/football_matches.csv")
print(df)
print(df.head(10))
```

```python
print(df.info())
```

```python
df ['date']=pd.to_datetime(df['date'])
print(df.info())
```

```python
print(df.isna().sum())
```

```python
df.duplicated().sum()
df.drop_duplicates(inplace=True)
```

```python
sns.boxplot(df)
plt.show()
```

```python
df.describe()
df.describe(include=object)
```

```python
print(df.corr())
```

```python
f=df['home_clearances'].corr(df['home_corners'])
print(f)
plt.scatter(df['home_clearances'],df['home_corners'])
plt.title("Correlation Matrix")
plt.xlabel('home_clearances')
plt.ylabel('home_corners')
plt.show()
```

```python
#matches played per season
matches_per_season=df["season"].value_counts().sort_index()
print(matches_per_season)
```

```python
#average number of goals per match
average_goals=df[['goal_home_ft','goal_away_ft']].sum(axis=1).mean()
print(f"Average goals per match: {average_goals:.2f}")
```

```python
#team with most and least goals
team_goals = df.groupby("home_team")[["goal_home_ft", "goal_away_ft"]].sum()
team_goals["total_goals"] = team_goals["goal_home_ft"] + team_goals["goal_away_ft"]
most_goals_team = team_goals["total_goals"].idxmax()
most_goals = team_goals["total_goals"].max()

least_goals_team = team_goals["total_goals"].idxmin()
least_goals = team_goals["total_goals"].min()

print(f"Team with most goals: {most_goals_team} ({most_goals} goals)")
print(f"Team with least goals: {least_goals_team} ({least_goals} goals)")
```

```python
#possession correlate with match outcomes
df["possession_diff"] = df["home_possession"] - df["away_possession"]
plt.figure(figsize=(8, 5))
sns.boxplot(x=df["result"], y=df["possession_diff"], palette="coolwarm")
plt.xticks([0, 1, 2], ["Home Win", "Draw", "Away Win"])
plt.xlabel("Match Outcome")
plt.ylabel("Possession Difference (Home - Away)")
plt.title("Possession Difference vs. Match Outcome")
plt.show()


```

```python
#home and away team performance
home_wins = (df['result'] == 0).sum()
draws = (df['result'] == 1).sum()
away_wins = (df['result'] == 2).sum()
total_matches = len(df)
home_win_percentage = (home_wins / total_matches) * 100
draw_percentage = (draws / total_matches) * 100
away_win_percentage = (away_wins / total_matches) * 100
print(f"Home Win Percentage: {home_win_percentage:.2f}%")
print(f"Draw Percentage: {draw_percentage:.2f}%")
print(f"Away Win Percentage: {away_win_percentage:.2f}%")
if home_win_percentage > away_win_percentage:
    print("Home teams generally perform better than away teams.")
else:
    print("Away teams perform equally or better thanhometeams.")
```

```python
#highest win percentages
home_matches = df.groupby("home_team")["home_team"].count()
away_matches = df.groupby("away_team")["away_team"].count()
total_matches = home_matches.add(away_matches, fill_value=0)
home_wins = df[df["result"] == 0].groupby("home_team")["home_team"].count()
away_wins = df[df["result"] == 2].groupby("away_team")["away_team"].count()
total_wins = home_wins.add(away_wins, fill_value=0)
win_percentage = (total_wins / total_matches) * 100
best_team = win_percentage.idxmax()
highest_win_pct = win_percentage.max()
print(f"Team with the highest win percentage: {best_team}")
print(f"Win Percentage: {highest_win_pct:.2f}%")
```

```python
# Total goals conceded by each team
home_goals_conceded = df.groupby("home_team")["goal_away_ft"].sum()
away_goals_conceded = df.groupby("away_team")["goal_home_ft"].sum()
total_goals_conceded = home_goals_conceded.add(away_goals_conceded, fill_value=0)
worst_defense_team = total_goals_conceded.idxmax()
most_goals_conceded = total_goals_conceded.max()
print(f"Team that conceded the most goals: {worst_defense_team}")
print(f"Total Goals Conceded: {most_goals_conceded}")
```

```python
# average cards per match per team per season
home_yellow_cards = df.groupby(["season", "home_team"])["home_yellow_cards"].sum()
away_yellow_cards = df.groupby(["season", "away_team"])["away_yellow_cards"].sum()
total_yellow_cards = home_yellow_cards.add(away_yellow_cards, fill_value=0)

home_red_cards = df.groupby(["season", "home_team"])["home_red_cards"].sum()
away_red_cards = df.groupby(["season", "away_team"])["away_red_cards"].sum()
total_red_cards = home_red_cards.add(away_red_cards, fill_value=0)

home_matches = df.groupby(["season", "home_team"])["home_team"].count()
away_matches = df.groupby(["season", "away_team"])["away_team"].count()
total_matches = home_matches.add(away_matches, fill_value=0)

avg_yellow_cards = total_yellow_cards / total_matches
avg_red_cards = total_red_cards / total_matches
cards_df = pd.DataFrame({
    "Avg Yellow Cards": avg_yellow_cards,
    "Avg Red Cards": avg_red_cards
}).reset_index()
plt.figure(figsize=(12, 6))
cards_df.groupby("season")[["Avg Yellow Cards", "Avg Red Cards"]].mean().plot(kind="bar", figsize=(12,6))
plt.title("Average Yellow and Red Cards per Team per Season")
plt.ylabel("Average Cards per Match")
plt.xlabel("Season")
plt.xticks(rotation=45)
plt.legend()
plt.show()
```

```python
#short on target and goal scored by teams
total_shots_on_target = df["home_shots_on_target"].sum() + df["away_shots_on_target"].sum()
total_goals = df["goal_home_ft"].sum() + df["goal_away_ft"].sum()
conversion = total_goals
missed_shots = total_shots_on_target - total_goals
labels = ["Goals Scored", "Shots on Target but Missed"]
sizes = [conversion, missed_shots]
colors = ["green", "red"]
plt.figure(figsize=(8, 8))
plt.pie(sizes, labels=labels, autopct="%1.1f%%", colors=colors, startangle=140, shadow=True)
plt.title("Relationship Between Shots on Target and Goals Scored")
plt.show()
```

```python
#matches ended in a draw
draw_matches=(df["result"]==1).sum()
print(f"total number of matches that ended in a draw: {draw_matches}")
```

```python
# team with the highest average away possession
avg_away_possession = df.groupby("away_team")["away_possession"].mean().sort_values(ascending=False)
best_team = avg_away_possession.idxmax()
highest_avg_possession = avg_away_possession.max()
print(f"Team with the highest average away possession: {best_team}")
print(f"Average Away Possession: {highest_avg_possession:.2f}%")
plt.figure(figsize=(12, 6))
avg_away_possession.head(10).plot(kind="bar", color="blue")
plt.title("Top 10 Teams with Highest Average Possession in Away Matches")
plt.ylabel("Average Possession (%)")
plt.xlabel("Team")
plt.xticks(rotation=45)
plt.show()
```

```python
# Correlation between possession difference and match result
df["possession_difference"] = df["home_possession"] - df["away_possession"]
correlation = df[["possession_difference", "result"]].corr().iloc[0, 1]
print(f"Correlation between possession difference and match result: {correlation:.2f}")
plt.figure(figsize=(8, 6))
sns.scatterplot(x=df["possession_difference"], y=df["result"], alpha=0.5)
plt.axhline(y=1, color="red", linestyle="--", label="Draw (1)")
plt.axhline(y=0, color="green", linestyle="--", label="Home Win (0)")
plt.axhline(y=2, color="blue", linestyle="--", label="Away Win (2)")
plt.title("Correlation Between Possession Difference and Match Result")
plt.xlabel("Possession Difference (Home - Away)")
plt.ylabel("Match Result (0 = Home Win, 1 = Draw, 2 = Away Win)")
plt.legend()
plt.show()
```

```python
#season with the most red cards
total_red_cards_per_season = df.groupby("season")[["home_red_cards", "away_red_cards"]].sum()
total_red_cards_per_season["total_red_cards"] = total_red_cards_per_season["home_red_cards"] + total_red_cards_per_season["away_red_cards"]
most_red_cards_season = total_red_cards_per_season["total_red_cards"].idxmax()
max_red_cards = total_red_cards_per_season["total_red_cards"].max()
print(f"Season with the most red cards: {most_red_cards_season}")
print(f"Total Red Cards: {max_red_cards}")
plt.figure(figsize=(12, 6))
total_red_cards_per_season["total_red_cards"].plot(kind="bar", color="red")
plt.title("Total Red Cards per Season")
plt.ylabel("Number of Red Cards")
plt.xlabel("Season")
plt.xticks(rotation=45)
plt.show()
```

```python
# Distribution  for match results
result_counts = df["result"].value_counts()
labels = {0: "Home Win", 1: "Draw", 2: "Away Win"}
result_counts.index = result_counts.index.map(labels)
plt.figure(figsize=(8, 8))
plt.pie(result_counts, labels=result_counts.index, autopct="%1.1f%%", colors=["green", "gray", "blue"], startangle=140)
plt.title("Distribution of Match Results")
plt.show()

```

```python
# team with more shots on target won
home_more_shots = df["home_shots_on_target"] > df["away_shots_on_target"]
away_more_shots = df["away_shots_on_target"] > df["home_shots_on_target"]
home_win_with_more_shots = ((df["result"] == 0) & home_more_shots).sum()
away_win_with_more_shots = ((df["result"] == 2) & away_more_shots).sum()
total_more_shots = home_more_shots.sum() + away_more_shots.sum()
win_percentage = ((home_win_with_more_shots + away_win_with_more_shots) / total_more_shots) * 100
print(f"Matches where a team had more shots on target: {total_more_shots}")
print(f"Matches won by the team with more shots on target: {home_win_with_more_shots + away_win_with_more_shots}")
print(f"Win Percentage when a team has more shots on target: {win_percentage:.2f}%")
```
